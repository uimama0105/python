##型別轉換

```python
x="must"
print(type(x))
y=10
print(type(y))
print(10*y)
print(10*x)
z="10"
print(10*z)
print(10*int(z))
print(10*float(z))
print(10*str(y))
print(10*eval("5*2"))
```
