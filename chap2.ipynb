#if else elif

```python
a=3
b=4
if a>b:  ##記得要冒號
  print("a>b")
elif a==b:
  print("a=b")  
else:
  print("a<b")
  
```

#成績評鑑

```python
grade=87
print("成績評鑑:")
if grade>=90:
  print("優等")
elif grade >=80:
    print("甲等")
elif grade >=70:
    print("乙等")
elif grade >=60:
    print("丙等")
else:
    print("不及格")
```

#雞兔同籠

```python
for chicken in range(0,36):
    rabbit=35-chicken
    if (2*chicken+4*rabbit) == 94:
        print("雞有",chicken,"隻,兔子有",rabbit,"隻")
        break
else:
    print("無解")
```

#While迴圈

```python
i=1
sum=0
while i < 11:
    sum+=i
    i+=1
    print(sum)
```

#本金10000元，定存年利率0.6%，請問多久銀行存款會翻倍

```python
x=10000
years=0
while x <= 20000:
    x=x*1.006
    years += 1
print(str(years), "年以後，存款加倍，為", int(x),"元")
```

#帳密判斷

```python
id="toyz"
pwd="12345"
while True:
    x=input("請輸入帳號：")
    y=input("請輸入密碼：")
    if (x==id and y==pwd):
        print("歡迎登入")
        break
    else:
        print("帳號或密碼輸入錯誤,請重新輸入")
```

#猜數字

```python
import random
number=random.randint(1,20)
while True:
    guess=int(input("請輸入1~20之間的數字-->"))
    if guess == number:
        print("你猜對了，數字是", number)
        break
    else:
        print("你猜錯了")
```
