##Tuple(元組)
tuple(元組)可視為不可改變的串列，tuple跟list很像，但是我們不能新增、刪除或更新tuple的元素。tuple的好處有：

1.佔用較少空間

2.可當作dictionary(字典)的鍵值

3.可當作函式引數

```python
a_tuple=('python','js','SQL')
print(a_tuple)
b_list=['go','c#','vb']
b_tuple=tuple(b_list)
print(b_tuple)
print(b_tuple[1])
```

```python
a_dict={0:'go',1:'c#',2:'vb'}
print(a_dict)
print(a_dict[0])
b_dict={'name':'tony','sex':'Male','sid':'B09090100'}
print(b_dict['sid'])
```
