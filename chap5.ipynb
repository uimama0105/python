#函式
-獨立的一組程式碼，可以重複使用

-通常可以有0-多個輸入參數，0~1個輸出

-函式輸入參數可以有預設值

```python
#傳統的程式寫法
import math
radius=eval(input("請輸入圓的半徑:"))
area= radius * radius * math.pi
print("圓的面積為: ",area)
```
