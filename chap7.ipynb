#錯誤處理
try:<br>
(需要檢測的程式碼)<br>
except:<br>
(錯誤時的處理方式)<br>
else:<br>
(如果沒有產生錯誤的處理方式)<br>
finally:<br>
(不管錯誤或執行正確，皆會進入的處理方式)<br>


```python
if a<b
  print(n)  
```

```python
  try:
    a=5
    b=10
    if a<b:
       print(n)
  except:
      print("錯誤產生")   
```
